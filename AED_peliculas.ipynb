```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('peliculas_a.csv')
```

```python
df.head()
```

```python
df.columns = ['edad','estudio','ciudad','vistas','accion','drama','romance','ciencia_ficcion','historia']
```

```python
df.head()
```

```python
df.info()
```

```python
features = df.drop(['accion','drama','romance','ciencia_ficcion','historia'], axis=1)
target = df[['accion','drama','romance','ciencia_ficcion','historia']]
```

```python
features_train, features_test, target_train, target_test = train_test_split(features,target,test_size=0.2,random_state=54221)
```

```python
print(features_train.shape)
print(target_train.shape)
print(features_test.shape)
print(target_test.shape)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(features_train,target_train)
predictions = knn.predict(features_test)
```

```python
genres = target.columns
for index,genre in enumerate(genres):
    score = accuracy_score(target_test[genre],predictions[:,index])
    print(f"Exactitud de predicción en el género {genre}: {score}\n")
```

```python
df_predictions = pd.DataFrame(data=knn.predict_proba(features_test)[4],columns=target.columns)
```

```python
target_test = features_test.reset_index(drop=True)
test = target_test.join(df_predictions)
```

```python
test.groupby(['edad','estudio','ciudad','vistas']).mean()[target.columns]
```
